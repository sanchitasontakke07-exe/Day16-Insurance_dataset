```python
#5
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot(2,2,1)
sns.histplot(df['price'])
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.ylabel("Count")
plt.subplot(2,2,2)
sns.histplot(df['launch_year'])
plt.title("Distribution of Launch Year")
plt.xlabel("Launch Year")
plt.ylabel("Count")
plt.subplot(2,2,3)
sns.histplot(df['engine_cc'])
plt.title("Distribution of Engine CC")
plt.xlabel("Engine CC")
plt.ylabel("Count")
plt.subplot(2,2,4)
sns.histplot(df['mileage_kmpl'])
plt.title("Distribution of Mileage")
plt.xlabel("Mileage")
plt.ylabel("Count")
plt.show()
```

```python
# 6
sns.countplot(x="brand", hue="stock_status", data=df)
plt.title("Brand vs Stock Status")
plt.show()

sns.countplot(x="launch_year", hue="stock_status", data=df)
plt.title("Launch Year vs Stock Status")
plt.show()

sns.countplot(x="owner_type", hue="stock_status", data=df)
plt.title("Seller Type vs Stock Status")
plt.show()
```

```python
#7
import matplotlib.pyplot as plt
import seaborn as sns
num_data = df[['price', 'engine_cc', 'mileage_kmpl', 'rating']]
corr = num_data.corr()
sns.heatmap(corr, annot=True, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#8
print("Average Price =", df["price"].mean())

print("Maximum Price =", df["price"].max())
print("Minimum Price =", df["price"].min())
print(df.groupby("brand")["price"].mean())
```

```python
#9
sns.boxplot(x="brand", y="price", data=df)
plt.title("Price by Brand")
plt.show()
sns.boxplot(x="stock_status", y="price", data=df)
plt.title("Price by Stock Status")
plt.show()
```

```python
#10
# 1. Average Price and Rating
print("Average Price:", df["price"].mean())
print("Average Rating:", df["rating"].mean())

# 2. Average Price by Brand
print("\nAverage Price by Brand:")
print(df.groupby("brand")["price"].mean())

# 3. Brand with Highest Number of Bikes
print("\nBrand with Highest Number of Bikes:")
print(df["brand"].value_counts().idxmax())

# 4. Summary
# Average bike price is ______.
# Average bike rating is ______.
# Bike prices vary across different brands.
# ________ brand has the highest number of bikes.
# The visualizations show the distribution of prices and brand-wise comparison.
```
